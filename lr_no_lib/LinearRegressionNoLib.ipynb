```python
# Do not edit the codes in this cell
# load required library
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
import numpy as np
# load dataset
X, y = load_diabetes(return_X_y=True)
X = X[:, 2]
# print ("X-----: ", X)
```

```python
# draw the raw data plot
# TODO
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('Data')
plt.scatter(X, y)
```

```python
# initialize theta
# TODO
theta0 = 0
theta1 = 0

print ("Theta Values ------ :")
print ("Theta0  ------ :", theta0)
print ("Theta1  ------ :", theta1)

def hypothesis(theta0, theta1, X):
    return theta0 + theta1*X
```

```python
# define cost function 
# TODO
data_length = len(y)
print ("Length of data ----- : ", data_length)
def cost_function(theta0, theta1, X, y):
  return (1/data_length) * (np.sum((hypothesis(theta0, theta1, X) - y)**2))
```

```python
# calculate cost by calling the function
# TODO
print ("Cost ---------- :", cost_function(theta0, theta1, X, y))
```

**Gradient Descent**

```python
# gradient descent to find the optimal fit
# TODO
## Since the initial error is very high we need to reduce it as much as possible to relatively higher epochs and learning rate is chosen.
## As per the figure given below loss is being reduced at very high rate for initial epochs and eventually reaching to a convergence after 2500 epochs.
## To balance the epochs a relatively higher alpha or learning rate value is taken by trial and error methodology to achieve a good fit which can be seen in the final figure.
def gradient_descent(theta0, theta1, X, y, epoch, alpha):
  cost = []
  counter = 0
  while (counter < epoch):
    hx = hypothesis(theta0, theta1, X)
    theta0 -= alpha*(sum(hx-y)/data_length)
    theta1 -= alpha * (np.sum((hx - y) * X))/data_length
    cost.append(cost_function(theta0, theta1, X, y))
    counter = counter + 1
  print ("Final Values ------- : ")
  print ("Theta0 ------- : ", theta0)
  print ("Theta1 ------- : ", theta1)
  print ("Cost   ------- : ", cost)
  print ("Cost   ------- : ", cost[-1])

  return theta0, theta1, cost
```

```python
# draw the cost change 
# TODO
def predict(theta0, theta1, X, y, epoch, alpha):
    theta0, theta1, cost = gradient_descent(theta0, theta1, X, y, epoch, alpha)
    return hypothesis(theta0, theta1, X), cost, theta0, theta1

y_predict, cost, theta0, theta1 = predict(theta0, theta1, X, y, 3000, 0.7)
plt.figure()
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Loss Function')
plt.scatter(range(0, len(cost)), cost)
plt.show()
```

```python
# draw the final linear model
# TODO
plt.figure()
plt.scatter(X, y, label = 'Original y')
plt.scatter(X, y_predict, label = 'predicted y')
plt.legend(loc = "upper left")
plt.xlabel("input feature")
plt.ylabel("Original and Predicted Output")
plt.show()
```
